```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

URL = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"

dados = pd.read_csv(URL)

dados_limpos = dados.dropna(subset=["Rua"])
dados_limpos
```

```python
dados_valor = pd.DataFrame()
dados_valor[["Moeda", "Valor_anun","forma"]] = dados["Valor"].str.split(expand = True)
dados_valor = dados_valor[dados_valor["forma"].isnull()]
dados_limpos["Valor"] = dados_valor["Valor_anun"].str.replace(".","").astype(float)
dados_limpos.dropna(subset = ["Valor"], inplace = True )
dados_limpos
```

```python
sns.set()
plt.figure(figsize=(10, 8))
ax = sns.histplot(data= dados_limpos["Valor"])
plt.xlim((-50, 10000000))
plt.xlabel("Custo dos imoveis em milhões")
plt.ylabel("Quantidade de imoveis")
ax.set_title("Preço de imoveis em São Paulo")
plt.show()
```
